## Data Types Breakdown

```python
# Integer

age = 10

print(type(age))
```

```python
# float

salary = 29999.99

print(type(salary))
```

```python
# string

name = 'ashish mehta'

print(type(name))
```

```python
# boolean

is_student = False

print(type(is_student))
```

```python
# list 

numbers = [1,2,3,4,5]

print(type(numbers))
```

```python
# Tuple

coordinates = (10, 20)

print(type(coordinates))
```

```python
# dictionary

student = {
    "name": "ashis bansal",
    "age" : 22   
}

print(type(student))
```

```python
# Set

unique_numbers = {1,2,3,4,5,6,6}

print(type(unique_numbers))
```
