# String Methods: Mastering Text Transformation

### Categories of methods:

- **Case conversion** –  upper() ,  lower() ,  swapcase() ,  capitalize() ,  title() 
- **Trimming & padding** –  strip() ,  lstrip() ,  rstrip() ,  center() ,  ljust() ,  rjust() ,  zfill() 
- **Searching & testing** –  find() ,  index() ,  count() ,  startswith() ,  endswith() ,  in  (operator)
- **Boolean checks** –  isalpha() ,  isdigit() ,  isalnum() ,  isspace() ,  isupper() ,  islower() 
- **Splitting & joining** –  split() ,  rsplit() ,  splitlines() ,  partition() ,  join() 
- **Replacing** –  replace() ,  translate() 
- **Formatting** –  format() ,  format_map() 

**Common mistake:** Forgetting that many methods (like split()) return a list, not a string, so chaining further string methods will fail.

```python
# Wrong 
result = "a, b, c".split(',').upper()
```

```python
# Correct
result = "a, b, c".upper().split(',')
result
```

- **Case conversion:**   
    - upper() ,  
    - lower() ,  
    - swapcase() ,  
    - capitalize() ,  
    - title() 


```python
text = "    python IS fun   "
print(text)

# Remove spaces and convert case
clean = text.strip()

print(clean.upper())
print(clean.lower())
print(clean.capitalize())
print(clean.swapcase())
print(clean.title())
```

- **Trimming & padding:**  
    - strip() ,  
    - lstrip() ,  
    - rstrip() ,  
    - center() ,  
    - ljust() ,  
    - rjust() ,  
    - zfill() 

```python
messy = "   hello   "

print(messy)
print(messy.strip())
print(messy.lstrip())
print(messy.rstrip())

# Padding to fixed with

num = "42"
print("\n")
print(num.center(6,"*"))
print(num.ljust(6,"-"))
print(num.rjust(6,"-"))
print(num.zfill(6))
```

- **Searching & testing:**  
    - find() ,  
    - index() ,  
    - count() ,  
    - startswith() ,  
    - endswith() ,  
    - in  (operator)

```python
sentence = "The quick brown fox jumps over the lazy dog."

print(sentence.find("fox"))
print(sentence.index("fox"))
print(sentence.count("o"))
print(sentence.startswith("The"))
print(sentence.endswith("dog."))
```

- **Boolean checks:**
    - isalpha() ,  
    - isdigit() ,  
    - isalnum() ,  
    - isspace() ,  
    - isupper() ,  
    - islower() 

```python
print("hello".isalpha())
print("123".isdigit())
print("hello123".isalnum())
print("  ".isspace())
print("HELLO".isupper())
print("hello".islower())
```

- **Splitting & joining:** 
    - split() ,  
    - rsplit() ,  
    - splitlines() ,  
    - partition() ,  
    - join() 


```python
data = "apple banana grapemango orange kiwi"

fruits = data.split(" ")
print(fruits)

print(data.split(" ",3)) #maximum 3 split

path = "/home/user/file.txt"
print(path.rsplit("/",1))

print(data.partition(' '))

print(data.splitlines())

restored = '-'.join(fruits)
print(restored)
```

- **Replacing:** 
    - replace() ,  
    - translate() 


```python
quote = "I like cats. cats are nice."

print(quote.replace("cats","dogs"))
print(quote.replace("cats","dogs",1))

trand_table = str.maketrans({'a':'@','o':'0'})
print("a hello world".translate(trand_table))
```

- **Formatting:** 
    - format() ,  
    - format_map() 

```python
template = "Name: {}, Age: {}"
print(template.format("Manoj",25))

print("{1} is {0}".format("old","this"))
print("{name} = {value}".format(name="x",value=42))

a=10
b=20
print(f"addition of a & b is {a+b}")
```
