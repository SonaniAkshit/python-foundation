### Indexing & Slicing: The Art of Extracting Data from Sequences

**Syntax rules**
- start  – first index to include (default 0)
- stop  – index to stop before (excluded, default = len(seq))
- step  – increment between indices (default 1, cannot be 0)

```python
# Simple indexing

s = "python"

print(s[0]) # first element of string
print(s[-1]) # last element of string
```

```python
# Slicing

print(s[0:])
print(s[:6])
print(s[::-1])
print(s[::2])
```

```python
nums = [0,1,2,3,4,5]
nums[1:4] = [100,200]
print(nums)
```

```python
# Delete Slice

nums[2:4] = []
print(nums)
```

```python
# Using slice() built-in

s = slice(1, 10, 2)

text = "Indexing & Slicing"
print(text[s])
```

```python
# Ellipsis slicing ( ... ) – used in NumPy/TensorFlow for multi‑dimensional arrays

import numpy as np
arr = np.arange(24).reshape(2,3,4)
print(arr,"\n")
print(arr[..., 1],"\n")   # all rows, all columns, only index 1 of last dimension
# Equivalent to arr[:, :, 1]
print(arr[:,:,1])
```
