# Python Tuples: The Immutable Power of Ordered Collections

A **tuple** is an **ordered, immutable, heterogeneous** sequence of objects.  
Key properties:

- **Ordered** – elements have fixed positions (indexing works).
- **Immutable** – cannot add, remove, or change elements after creation.
- **Heterogeneous** – can mix types: `(1, "hello", 3.14)`.
- **Hashable** – if all elements are hashable, the tuple itself can be used as a dict key or set element.
- **Indexable & slicable** – supports all indexing/slicing operations (returns new tuples).
- **Iterable** – can be looped over.
- **Allows duplicates** – `(1, 1, 2)` is fine.


```python
# Demonstrating properties

# Heterogeneous 
t = (10, "Python", 3.14)
print(type(t))

# ordered
print(t[0])

# immutable
# t[1] = "java" #TypeError: 'tuple' object does not support item assignment

# Hashable - can be dict key
d = { (1,2): "point"}
print(d[(1,2)])

# Indexable & slicable

print(t[0])
print(t[1])

print(t[0:2])
print(t[::2])
print(t[::-1])

# Iterable – can be looped over.
i = 0
for item in t:
    i+=1
    print(f"{i}:{item}")
    

# Allows duplicates

t2 = (1,1,2,2)
t2
```

### Creating Tuples

```python
# 1. Literal (parentheses)
empty = ()
print(empty)
single = (5,)
print(single)
multi = (1,2,3)
print(multi)
mixed = (1,"python",99.8)
print(mixed)

# 2. tuple() constructor
from_str = tuple("abc")
print(from_str)
from_list = tuple([1,"java",True])
print(from_list)
from_range = tuple(range(1,11))
print(from_range)

# 3. Without parentheses - tuple paking
packed = 1,2,3
print(packed)

# 4. Comprehension? Tuples don't have comprehension – use generator with tuple()
squares = tuple(x**2 for x in range(1,6))
print(squares)

# 5. using multiplication

zeros = (0,) * 5
print(zeros)

# unpacking / star expressions
combined = (1,2) + (3,4)
print(combined)
```

###  Advanced Concepts

```python
# Tuple Unpacking

# basic unpacking
a, b, c, = (1,2,3)
print(f"a={a},b={b},c={c}")

# using * for rest
first, *rest = (10,20,30,40,50)
print(f"first={first},rest={rest}")

*head, last = (100,200,300)
print(f"head={head},last={last}")

# Swapping variables (tuple packing/unpacking)

x,y = 5,10
print(f"before swap:x={x},y={y}")
x,y = y,x
print(f"after swap:x={x},y={y}")

# Returnig multiple values from function

def get_user():
    return (101, "alice", "alice@example.com")

id, name, email = get_user()

print(f"id={id},name={name},email={email}")
```

```python
# Named tuples (from collections modeule)

from collections import namedtuple

Point = namedtuple('Point', ['x', 'y'])
p = Point(10,20)

print(p.x, p.y)
print(p[0])
# p.x = 30 AttributeError: can't set attribute
```

### Mathematical / Special Operations


Tuples support the same sequence operations as lists, but no element‑wise arithmetic. However, they are often used in **mathematical contexts** to represent fixed vectors, coordinates, or parameters.

- **Concatenation**: `(1,2) + (3,4) = (1,2,3,4)`
- **Repetition**: `(1,2) * 2 = (1,2,1,2)`
- **Comparison**: Lexicographic order: `(1,2) < (1,3)` → `True`


```python
print((1,2) < (2,1))
print((1,2) == (1,2))

print((1,2) + (3,4))
print((1,)*5)

# Using tuples for multiple assignment in loops
pairs = [(1,'a'),(2,'b'),(3,'c')]
for num,char in pairs:
    print(num, char)
```

```python
# Example 3: Using tuple as a light-weight struct

# Representing RGB colors

colors = {
    "red": (255, 0, 0),
    "green": (0, 255, 0),
    "blue": (0, 0, 255)
}

def mix(color1, color2):
    return tuple((c1+c2) // 2 for c1, c2 in zip (color1, color2))

print(mix(colors["red"], colors["blue"]))
```
