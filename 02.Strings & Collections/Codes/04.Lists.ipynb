# Important List Methods

```python
# append()

numbers = [10, 20, 30, 40]

numbers.append(50)

numbers
```

```python
strings = ["rahul", "mayank", "sumit"]

strings.append("harsh")

strings
```

```python
# insert()

numbers.clear()

numbers = [10, 20, 30]

numbers.insert(1,33)
numbers
```

```python
strings.clear()

strings = ["rahul","harsh"]

strings.insert(1, "jethalal")
strings
```

```python
# remove()
numbers = [10,20,30,40]

numbers.remove(20)
numbers
```

```python
strings = ["rahul","harsh"]

strings.remove("rahul")
strings
```

```python
# pop()

numbers = [10,20,30,40,50,60,70,80]

numbers.pop(1)
# numbers
```

```python
strings = ["ram", "rahul", "radha", "chetan"]

strings.pop(3)

# strings
```

```python
# sort()

numbers = [10,90,87,45,23,65,77,23,54,11,89,00,76,35]
numbers.sort()
numbers
```

```python
strings = ["vaibhav", "alka", "mahek", "mayank"]

strings.sort()

strings
```

```python
# reverce()

numbers = [10,90,87,45,23,65,77,23,54,11,89,00,76,35]
numbers.reverse()
numbers
```

```python
strings = ["vaibhav", "alka", "mahek", "mayank"]

strings.reverse()

strings
```

```python
# extend()
lis = [10,20,30,40]

lis.extend(
    [500,600]
)

print(lis)
```

```python
# clear()
lis.clear()
print(lis)
```

```python
# copy()
lis = [10,20,30,40]
lis2 = lis.copy()
```

```python
# count()
lis = [10,20,30,40,10,30,10]
lis.count(10)
```

# Looping Through Lists

```python
students = ["vishva", "rohan", "manav", "vaibhav", "alka", "mahek", "mayank"]
i = 0
for student in students:
    i+=1
    print(f"{i} - {student}") # return students with index
```

```python
# Ex 2: Shoping Cart

cart = []

cart.append("laptop")
cart.append("mouse")
cart.append("keybord")

print(cart)
```

```python
# Ex 3: Dynamic Shoping Cart

catagory = input("Enter Catagory of item: ")
howmany = int(input("How many items want to add to cart?: "))

items = []

for i in range(1, howmany + 1):
    take = input(f"Add {catagory} {i} to Add in Cart")
    items.append(take)

j = 0

if len(items) == 0:
    
    print("Your Cart is Empty...")

else:    
    
    print("---Your Cart Items ----")
    for item in items:
        j += 1
        print(f"{catagory} {j}: {item}")
```
