# Python Lists: The Versatile Dynamic Array


A **list** is a **mutable, ordered, heterogeneous, dynamic** collection of objects.  
Key properties:

- **Ordered** – elements have a defined position (index starting at 0).
- **Mutable** – you can change, add, or remove elements in‑place.
- **Heterogeneous** – can contain different types: `[1, "hello", 3.14, [2,3]]`.
- **Dynamic** – size can change automatically (no fixed capacity).
- **Indexable & slicable** – supports all indexing/slicing operations.
- **Iterable** – can be used in loops.
- **Allows duplicates** – multiple same values are fine.

```python
# Demonstrating properties

# Heterogeneous
lst = ["python", 1, 3.14, True, [1,2]]

print(type(lst))
print(lst)
```

```python
# Ordered
print(lst[0])
print(lst[1])
```

```python
lst[1] = 10
print(lst[1])
print(lst)
```

```python
# Dynamic

lst = ['python', 1, 3.14, True, False, [1,2,3,4,5]]
print(lst)
```

```python
# Indexable & slicable

print(lst[0])
print(lst[1])
print(lst[2])

print(lst[3:5])
```

### Creating List

```python
# 1. Literal (most comon)

empty = []
nums = [1, 2, 3, 4]
mix = [1, "two", 3.0]

print(type(empty),type(nums),type(mix))
print(empty,nums,mix)
```

```python
# 2. list() constructor

from_str = list("hello")
from_tuple = list((1,2,3))
from_range = list(range(5))

print(from_str)
print(from_tuple)
print(from_range)
```

```python
# 3. list comprehension

squares = [x*x for x in range(1,6)]
print(squares)
```

```python
# 4. Using multiplication (repetition)

zeros = [0] * 5
print(zeros)

matrix = [[1]*3 for _ in range(4)] # 4D matrix
print(matrix)
```

```python
# 5. Splitting a string

words = "hello world".split()

print(words)
```

```python
# 6. Unpacking / star expressions (Python 3.5+)

combined = [1,2] + [3,4]
print(combined)
```

### Operations / Methods

```python
lst = [1, 2]

# add elements
lst.append(3)
print(f"append:{lst}")

lst.extend([4,5])
print(f"extend:{lst}")

lst.insert(6,6)
print(f"insert:{lst}")


# Remove elements
lst.remove(6)
print(f"remove:{lst}")

lst.pop(0)
print(f"pop:{lst}")

# find
lst = ["a", "b", "c", "d", "a"]
print(f"index:{lst.index('a')}")

print(f"count:{lst.count('a')}")

# sortig
lst = [1, 2, 10, 3, 6, 7, 4, 5, 8, 9]
lst.sort()
print(f"sort:{lst}")

lst.reverse()
print(f"reverse:{lst}")

xyz = lst.copy()
print(f"copy:{xyz}")

xyz.clear()
print(f"clear:{xyz}")
```

### Advanced Concepts

```python
# List slicing (as learned earlier)

nums = [0, 1, 2, 3, 4, 5]
print(nums[1:4])
print(nums[::2])
print(nums[::-1])
```

```python
# slice assignment (powerful mutation)

lst = [1,2,3,4,5]
lst[1:4] = [20,30]
print(lst)

lst[2:2] = [25,26]
print(lst)

lst[::2] = [10,20,30]
print(lst)
```

```python
# List comprehensions 

# sequence of even numbers

enven_sq = [x**2 for x in range(10) if x % 2 == 0]
print(enven_sq)

# nested comprehension

matrix = [[1,2],[3,4]]
flat = [num for row in matrix for num in row]

flat
```

```python
# Packing & unpacking

a, b, c = [1,2,3]
print(a)
print(b)
print(c)
```

```python
first, *rest = [10,20,30,40]
print(first)
print(rest)
```

```python
*head, last = [1,2,3]
print(head)
print(last)
```
