## Indexing & Slicing

```python
name = "raghav"

# first Charactor
print(name[0])

# last Charactor
print(name[-1])
```

```python
# Extract domain

email = "user@gmail.com"

domain = email[email.index("@") + 1:]

print(domain)
```

```python
# Extract File Extension

filename = "model_weights.pkl"

extension = filename[-4:]

print(extension)
```
