```python
# upper()

str = "hello"

print(f"original string: {str}")
str.upper()
```

```python
# lower()

str1 = 'HELLO'

print(f"original string : {str1}")
str1.lower()
```

```python
# title()

str2 = 'hey! my name is akshit i m learning python.'

print(f"original string: {str2}")

str2.title()
```

```python
# capitalize()

str3 = str2

print(f"original string: {str3}")
str3.capitalize()
```

```python
# strip()

str4 = "        hello       "

print(f"original string: {str4}")
str4.strip()
```

```python
# lstrip()

str5 = "    hello"

print(f"original string: {str5}")
str5.lstrip()
```

```python
#  rstrip()

str6 = "hello       "

print(f"original string: {str6}")
str6.rstrip()
```

```python
# replace()

str7 = "I love java"

print(f"original string: {str7}")
str7.replace("java", "Python")
```

```python
# find()

str8 = str7

print(str8)

str8.find("java") # return index of first occurence
```

```python
# count()

str9 = "Return the number of non-overlapping occurrences of substring sub in string S[start:end]. Optional arguments start and end are interpreted as in slice notation."

print(str9)

str9.count("a")
```

```python
# startswith()
str10 = str9

print(str10)
str10.startswith("Return")
```

```python
# endswith()
str11 = str10

print(str11)
str11.endswith("notation.")
```

```python
# split()
str12 = str10

print(f"STRING: {str12}")
split = str12.split()

print("\n",split)
```

```python
# join()
str13 = split

print(f"split list: {str13}")

join = " ".join(str13)
join
```

```python
# isalpha()

str14 = 'abc'
print(str14.isalpha())

str15 = 'abc123'
print(str15.isalpha())

```

```python
# isdigit()

str16 = 'abc123'
print(str16.isdigit())

str17 = '123'
print(str17.isdigit())

```

```python
# isalnum()

str18 = 'abc123'
print(str18.isalnum())

str19 = 'abc'
print(str19.isalnum())

str20 = '123'
print(str20.isalnum())

```
