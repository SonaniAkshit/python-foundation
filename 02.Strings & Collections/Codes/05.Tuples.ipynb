## Creating Tuples

```python
# Empty Tuple

empty_tuple = ()

type(empty_tuple)
```

```python
# with values

student = ("akshit", 21, 64.3)
print(type(student))
print(student)
```

```python
# Tuple Without Parentheses

student = "akshit", 21, 64.3

print(type(student))
print(student)
```

```python
# Single Element Tuple

num = (10)
print(type(num))
```

```python
string = ("akshit")
print(type(string))
```

```python
# Single Element Tuple

num = (10, )
print(type(num))
```

```python
string = ("akshit", )
print(type(string))
```

```python
# Tuple Immutability

numbers = (10, 20, 30)

numbers[0] = 100
```

## Tuple Packing Unpacking

```python
# Tuple Packing

student = "keshav", 21, 85  # Equivalent to: student = ("keshav", 21, 85)

print(student)
print(type(student))
```

```python
# Tuple Unpacking

name, age, marks = ("keshav", 21, 89)

print(name)
print(age)
print(marks)
```

## Multiple Return Values

Python actually returns tuples.

```python
def get_student():
    return "Keshav", 21
data = get_student()

print(data)
print(type(data))
```

```python
# Unpacking Return Values

def get_student():
    return "keshav", 21
name, age = get_student()

print(name)
print(age)
```

## Memory Behavior

```python
import sys

my_list = [1, 2, 3]
my_tuple = (1, 2, 3)

print(f"List memory: {sys.getsizeof(my_list)} bytes")
print(f"Tuple memory: {sys.getsizeof(my_tuple)} bytes")
```

```python
# Student Record System

student = (
    "Keshav",
    21,
    "Machine Learning",
    98.99
)

print(f"Name: {student[0]}")

print(f"Age: {student[1]}")

print(f"Course: {student[2]}")

print(f"Marks: {student[3]}")
```

Tuple Benefit:

- Record remains unchanged.
- Prevents accidental modification.
