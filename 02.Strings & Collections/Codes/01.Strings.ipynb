##  String Coding Examples

```python
name = "ram"
print(name)
print(type(name))
```

```python
# Accessing Characters

word = "python"

print(f"Index [0]: Char : {word[0]}")
print(f"Index [1]: Char : {word[1]}")
print(f"Index [2]: Char : {word[2]}")
print(f"Index [3]: Char : {word[3]}")
print(f"Index [4]: Char : {word[4]}")
print(f"Index [5]: Char : {word[5]}")
```

```python
# Negative Indexing

word = "python"

print(f"Index [-1]: Char : {word[-1]}")
print(f"Index [-2]: Char : {word[-2]}")
print(f"Index [-3]: Char : {word[-3]}")
print(f"Index [-4]: Char : {word[-4]}")
print(f"Index [-5]: Char : {word[-5]}")
print(f"Index [-6]: Char : {word[-6]}")
```

```python
# Slicing

text = "MachineLearning"

print(text[0:7])
print(text[7:])
```

```python
# Reverse String

text = "MachineLearning"

print(text[::-1])
```

```python
# Email Validation

email = "hello@gmail.com"

if "@" in email:
    print("Valid")
else:
    print("Invalid")
```

```python
# Cleaning User Input

name = "        raghav      "

clean_name = name.strip().title()

print(clean_name)
```

###  Important String Methods

```python
lower = "HELLO"

print(lower.lower())
```

```python
upper = "hello"

print(upper.upper())
```

```python
strip = "       hello       "

print(strip.strip())
```

```python
replace = "Machine Learning"

print(replace.replace("Machine", "Deep"))
```

```python
split = "Machine Learning"

print(split.split())
```

```python
title = 'machine learning'

print(title.title())
```

```python
join = ["machine", "learning"]

result = " ".join(join).title()
print(result)
```

```python
find = "Hey i'm curruntly learning on machine learning"

print(find.find("machine"))
```

```python
prefix = "Machine Learning"

print(prefix.startswith("M"))
```

```python
suffix = "Machine Learning"

print(suffix.endswith("g"))
```
