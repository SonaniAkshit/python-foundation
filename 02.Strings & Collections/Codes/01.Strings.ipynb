# Strings: The Immutable Sequence of Characters


### Key properties:

- Ordered – characters have a fixed position.
- Immutable – cannot be changed in‑place.
- Iterable – can loop over characters.
- Indexable – access by position:  s[0] ,  s[-1] .
- Hashable – can be used as dictionary keys or set elements.
- Supports concatenation and repetition:  +  and  * 

**Demonstration of core properties:**

```python
s = "Python"
```

```python
# Ordered & indexable

print(s[0])
print(s[-1])
```

```python
# Immutable – this will raise an error

try:
    s[0] = "J"
except TypeError as e:
    print(f"Error: {e}")
```

```python
# Iterable

for ch in s:
    print(ch)
```

```python
# Hashable – works as dict key

d = {s: "hello"}

print(d[s])
```

```python
# Concatenation and repetition:  +  and  * .

s1 = "Python"
s2 = "Java"

print(f"{s1*5}")
print(f"{s1+s2}")
```

### Creating Strings


```python
# 1. Literals (most common)

s1 = 'hello'
s2 = "world"
s3 = '''Multiline
string'''
s4 = """One more
Multiline
String"""

print(s1)
print("\n",s2)
print("\n",s3)
print("\n",s4)
```

```python
# 2. Using str() constructor

s5 = str(123)
s6 = str(6.43)
s7 = str([1,2,3])
s8 = str("hello")

print(type(s5))
print(type(s6))
print(type(s7))
print(type(s8))

print(f"{s5}\n{s6}\n{s7}\n{s8}")
```

```python
# 3. From bytes with encoding

b = b'hello'
s9 = b.decode('utf-8')
s9
```

```python
# 4. Comprehension (using join and generator)

s10 = ' '.join(str(i) for i in range(5))
s10
```

```python
# 5. Using string multiplication

s11 = "ha"
print(s11 * 5)
```

```python
# 6. Empty string

empty = ""
empty
```

### Common mistakes


```python
# Mistake 1: mixing quotes incorrectly
# s = 'It's fine'   # SyntaxError
# Correct:
s = "It's fine"
s
```

```python
# Mistake 2: forgetting that strings are immutable
# s = "hello"
# s[0] = "H"         
# TypeError
```

```python
# Mistake 3: using + in a loop (performance killer)
result = ""
for ch in ["a", "b", "c"]:
    result += ch   # O(n²) – creates many intermediate strings
result   
# Better: ''.join(["a", "b", "c"]
```

###  Advanced Concepts

```python
# Slicing
# [start:stop:step]  – returns a new string.

s = "0123456789"

print(s[2:7])
print(s[:5])
print(s[5:])
print(s[::2])
print(s[::-1])
```

```python
# Packing & Unpacking

a,b,c = "XYZ"
a
```

```python
first, *rest = "Python"
first
```

```python
rest
```

```python
# String Comprehension (via generator)

s = ''.join(str(x) for x in range(10) if x % 2 == 0)
print(s) 
```

```python
# Nested structures (list of strings, etc.)

matrix_str = ["abc", "def", "ghi"]
# Accessing characters
print(matrix_str[1][2])
```

```python
# Raw strings (escape sequences ignored)

path = r"C:\Users\Name"   # no need to double backslashes
print(path)
```

```python
# F strings

a = 10
b = 20
result = f"addition of a+b is: {a+b}"
result
```

```python
# f‑strings with expressions

a, b = 5, 3
print(f"{a} + {b} = {a + b}")
print(f"{a:.2f}")               
```

```python
# Membership:  in  and  not in 

print("Py" in "Python")
print("X" not in "Hi")
```

```python
# Comparison: Lexicographic order based on Unicode code points.

print("apple" < "banana")
```

```python
print("Apple" < "apple")
```
