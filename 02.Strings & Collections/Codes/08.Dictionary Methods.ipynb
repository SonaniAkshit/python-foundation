**Dictionary Methods**

## 1. Introduction

Python dictionaries (`dict`) are mutable, unordered (insertion-ordered since Python 3.7) collections of key-value pairs that provide average O(1) lookup time. They exist because efficient key-based access is fundamental for fast data retrieval and organization. In ML engineering, dictionaries are indispensable for configuration management, feature mapping, hyperparameter storage, and intermediate data processing pipelines.



**`get(key, default=None)`**  
Purpose: Safe key access without raising KeyError.  
Syntax: `d.get(key, default)`  
Returns: Value or default. 

```python
config: dict[str, int] = {"epochs": 100}

print(config.get("epochs"))
print(config.get("batch_size")) # return default : None
print(config.get("batch_size", 32)) # return value: 32
```

**`keys()`, `values()`, `items()`**  
Purpose: View keys/values/pairs (memory-efficient views).

```python
data = {
    "feature1" : 0.8,
    "feature2" : 0.3
}

print(data.keys())
print(list(data.values()))
print(list(data.items()))

# all in one
print("\n")
for key, value in data.items():
    print(f"{key}: {value}")
```


**`update([other])`**  
Purpose: Merge another dict or iterable of pairs.  

```python
params = {"lr": 0.001}
params.update({"batch_size": 64, "lr": 0.01})  # overwrites
print(params)
```

**`pop(key, default)`**  
Purpose: Remove and return value (with optional default).  


```python
metrics = dict(
    loss = 0.45,
    acc = 0.92
)
print(f"before pop: {metrics}")

loss = metrics.pop("loss")
print(f"poped element: {loss}")

print(f"After pop: {metrics}")
```

**`setdefault(key, default)`**  
Purpose: Insert default if key missing, then return value.  
**`popitem()`**: Remove and return last inserted pair (LIFO).  
**`clear()`** and **`copy()`**: Self-explanatory.


```python
# setdefault(key, default)

student = {
    "name": "alice"
}

student.setdefault("age", 22)

print(student)
```

```python
# popitem()

data = {
    "a": 1,
    "b": 2,
    "c": 3
}

item = data.popitem()

print(item)
print(data)
```

```python
# clear()
# Removes everything from the dictionary.

data = {
    "a": 1,
    "b": 2,
    "c": 3
}

data.clear()

print(data)
```

```python
# copy
student = {
    "name": "alice",
    "age": 22,
    "gender": "female",
    "course": {
        "cs": "bca",
        "ed": "bed"
    }
}

student_copy = student.copy()

print(student_copy)
```
