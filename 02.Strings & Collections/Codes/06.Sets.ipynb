### Creating Sets

```python
# Method 1

fruits = {"apple", "banana", "orange"}
print(fruits)
print(type(fruits))
```

```python
# Method 2: Using set()

fruits = set(["apple", "banana", "orange", "apple"])
print(fruits)
```

```python
# empty set

data = {} #wrong

print(type(data))
```

```python
data = set() #Correct
print(type(data))
```

```python
# **Adding Elements**
# ======================

# add()

numbers = {1,2,3,4,5}
numbers.add(6)
numbers
```

```python
# **Adding Multiple Elements**
# ======================

# update()

numbers = {1,2,3,4}

numbers.update([5,6,7,8])

numbers
```

```python
# **Removing Elements**
# ======================

# remove()

numbers = {1, 2, 3, 4, 5}

numbers.remove(5)

numbers
```

```python
# discard()

numbers = {1,2,3,4}

numbers.discard(5)

numbers
```

```python
# pop()

numbers = {1,2,3,4}

numbers.pop()

numbers
```

```python
# clear()

numbers = {1,2,3,4,5}

numbers.clear()

numbers
```

```python
# Membership Testing
# ===================

users = {"akshit", "reahul", "jay"}

print("akshit" in users)
```

### Set Operations

```python
A = {1, 2, 3, 6}
B = {3, 4, 5, 1}
```

```python
# Union
# Combines both sets.

print(A | B)
```

```python
# Difference
# Values in A not in B.

print(A - B)
```

```python
# Difference
# Values in B not in A.

print(B - A)
```

```python
# Intersection
# Common values.

print(A & B)
```

```python
# Symmetric Difference
# Values not common.

print(A ^ B)
```

###  Frozen Sets

Normal sets are mutable.

Sometimes we need immutable sets.

```python
data = frozenset([1, 2, 3])
```

Cannot modify.

Useful for:

- Dictionary keys
- Caching
- Configuration values

## Practical Examples

```python
# Remove Duplicates

data = [1, 2, 2, 3, 3, 4, 4, 5]

unique = set(data)

data = list(unique)

data
```

```python
# Unique Visitors

numbers = int(input("How many visitor right now?: "))

visitors = set()

for i in range(1, numbers+1):
    name = input("Enter visitor name: ")
    visitors.add(name)
    print(f"add visitor {i} name: {name}")

print(f"Unique Visitors: {visitors}")
```
