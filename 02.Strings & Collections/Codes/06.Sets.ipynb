# Python Sets: The Unordered Collection of Unique Elements

A **set** is an **unordered, mutable, iterable collection of unique, hashable objects**.  
Key properties:

- **Unordered** – elements have no index; you cannot access `s[0]`.
- **Mutable** – you can add and remove elements.
- **Unique** – duplicates are automatically eliminated.
- **Hashable elements** – every element must be immutable and hashable (strings, numbers, tuples of hashables, but not lists or dicts or other sets).
- **Fast membership** – `x in s` is O(1) average.
- **Iterable** – can loop over elements (order is arbitrary but deterministic within a run).
- **Not indexable or slicable** – because there is no order.


```python
s = {1, 2, 3, 3, 4}
print(s)
```

```python
# Unordered 
s[0]
```

```python
# Mutable

s.add(5)
print(s)
```

```python
# Fast membership 
6 in s
```

```python
fruits = {"apple", "banana", "cherry", "mango", "apple"}
print(fruits)
i = 0
for fruit in fruits:
    i += 1
    print(i, fruit)
```

### Creating Sets

```python
# 1. Set literal

empty = set()
print(empty)
numbers = {1,2,3,3,4,5}
print(numbers)
mixed = {1, "hello", 3.14}
print(mixed)

# 2. set() constuctor from iterable
from_list = set([1,2,3,4])
print(from_list)
from_string = set("hello i m student")
print(from_string)
from_range = set(range(1,11))
print(from_range)

# 3. Set comprehension

squers = {x**2 for x in range(1,6)}
even_squers = {x**2 for x in range(1,11) if x % 2 == 0}
print(squers)
print(even_squers)

# 4. Using set() on a generator

uniqe_chars = set(ch for ch in "hello" if ch.isalpha())
print(uniqe_chars)

# 5. from dictionary keys

d = {'a':1, 'b':2, 'c':3}

keys_set = set(d)

print(keys_set)
```

### Core Operations / Methods

```python
set1 = {1,2,3,4,5}

# add
set1.add(3) #Adds element if not present
print(set1)

# remove
set1.remove(3) #Removes; raises KeyError if missing
print(set1)

# discard
set1.discard(3) #Removes if present, no error
print(set1)

```

```python
# pop
set1.pop() # Removes and returns arbitrary element
```

```python
# clear
set1.clear() # clear whole set and return 'set()' as output
print(set1)

# copy 
{1,2}.copy() 
```

```python
A = {1,2,3}
B = {3,4,5}

# Union
print(A | B) # All elements from both, Combine sets

# intersection
print(A & B) # Common elements, Find overlaps

# Difference
print(A - B) # Elements in A but not B, Exclude elements

# Symmetric Difference
print(A <= B) # Checks All elements of A in B, Test containment

# Superset check
print(A >= B) # A is subset of B, Test containment

# Disjoint
print(A.isdisjoint(B)) # No common elements, Check separation
```

### Advanced Concepts

### Frozenset – immutable set

Frozensets are hashable and can be used as dictionary keys or elements of another set.


```python
fs = frozenset([1,2,3])
# fs.add(4) AttributeError: 'frozenset' object has no attribute 'add'

d = {fs: "value"}
print(d)

s = {frozenset([1,2]), frozenset([3,4])}
print(s)
```

```python
# Set comprehensions with conditions

words = ["hello", "world", "python", "set"]
unique_lengths = {len(w) for w in words}
print(unique_lengths)
long_words = {w for w in words if len(w) > 4}
print(long_words)
```

```python
# Updating sets in‑place with operations

s = {1,2}
s |= {2,3,4} #add
print(s)

s &= {1,2,3} #remove remain
print(s)

s -= {3} #remove selected
print(s)

s ^= {4,5}
print(s)
```

```python
# Using update, intersection_update, etc.

s1 = {1,2,3}
s1.update([3,4,5])
print(s1)

s1.intersection_update({2,3,4})
print(s1)
```

```python
# Set as a fast membership test in loops

allowed = {"read", "write", "execute"}
user_perms = ["read", "delete", "write"]

for perm in user_perms:
    if perm in allowed:
        print(f"{perm} is allowed")
```

### Mathematical / Special Operations

Sets directly implement **set algebra** operations:

| Operation | Python | Mathematical notation |
|-----------|--------|----------------------|
| Union | `s1 \| s2` | A ∪ B |
| Intersection | `s1 & s2` | A ∩ B |
| Difference | `s1 - s2` | A \ B |
| Symmetric difference | `s1 ^ s2` | A Δ B |
| Subset | `s1 <= s2` | A ⊆ B |
| Proper subset | `s1 < s2` | A ⊂ B |
| Superset | `s1 >= s2` | A ⊇ B |
| Proper superset | `s1 > s2` | A ⊃ B |

These allow you to compute **Jaccard similarity**: `|A ∩ B| / |A ∪ B|`


```python
s1 = {1,2}
s2 = {2,3}

print(s1 | s2)
print(s1 & s2)
print(s1 - s2)
print(s1 ^ s2)
print(s1 <= s2)
print(s1 < s2)
print(s2 >= s2)
print(s1 > s2)
```

```python
def jaccard (A, B):
    return len(A & B) / len (A | B)

A = {1,2,3,4}
B = {3,4,5,6}

print(jaccard(A, B))
```

### Practical Examples

```python
# Example 1: Finding unique visitors from logs
log_ips = [
    "192.168.1.1", "10.0.0.2", "192.168.1.1", "172.16.0.1", "10.0.0.2"
]
unique_ips = set(log_ips)
print(unique_ips)
print(f"unique visitors: {len(unique_ips)}")
```

```python
# Example 2: Tag intersection for content recommendation

user_interests = {"python", "data science", "machine learning", "sql"}
article_tags = {"machine learning", "deep learning", "python", "nlp"}

common = user_interests & article_tags
print(f"Recommended because of : {common}")

# Suggest tags from article not in user interests
new_tags = article_tags - user_interests
print(f"New topics to explore: {new_tags}")
```

```python
# Example 3: Deduplicating a list while preserving order (Python 3.7+)

def unique_preserve_order(seq):
    seen = set()
    return [x for x in seq if not (x in seen or seen.add(x))]

items = [3,1,2,1,2,3,4]
print(unique_preserve_order(items))
```
