**4. Develop a system that converts temperature from Celsius to Fahrenheit using type conversion and arithmetic operators.**

```python
celsius = int(input("Celsius: "))

fahrenheit = (celsius * 9) / 5 + 32

print(f"Fahrenhit: {int(fahrenheit)}°F")
```

**6. Write a program that checks whether a number entered by the user is even or odd using comparison and modulus operators.**

```python
number = int(input("Enter number: "))

print(f"Number: {number}")
if number % 2 == 0:
    print("Even")
else:
    print("Odd")
```

**8. Create a marks analysis program that accepts marks as string input, converts them into integers, and calculates the average score.**

```python
python = input('mark of python: ')
java = input('mark of java: ')
cpp = input('mark of c++: ')
c = input('mark of c: ')

total = int(python) + int(java) + int(cpp) + int(c)
avg = total / 4

print("Marksheet")
print("___________")

print(f"Python: {python}")
print(f"java: {java}")
print(f"C++: {cpp}")
print(f"C: {c}")

print("----------------")
print(f"Total Score: {total}")
print(f"Average Score: {avg}")
```
