### Beginner Example

### Integer Input


```python
age = int(input("Enter your age: "))
print("your age", age)
```

```python
print(age + 5)
```

Why int()?

Because input() returns a string.

int() converts it into an integer.

### Float Input

```python
salary = float(input("Enter your Salary: "))
print("your Salary", salary)
```

### Multiple Inputs


```python
a, b = input("Enter two values: ").split()
print(a)
print(b)
print(a+b)
```
