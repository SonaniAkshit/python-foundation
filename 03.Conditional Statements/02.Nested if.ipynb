# **Nested Conditional Statements**



## 1. Introduction

Nested conditionals mean placing one `if-elif-else` structure inside another. This allows handling multiple levels of decision-making where the inner decision depends on the outer one being true.m

## 2. Intuition

Imagine a **multi-stage security checkpoint** at an airport:

- Outer `if`: "If you have a boarding pass..."
  - Inner `if`: "...and your ID matches, then check luggage rules."
    - Even deeper: "...and if you're carrying liquids, apply additional screening."

Each level only gets evaluated if the previous gate opens. This mirrors how nested `if`s work — inner blocks are unreachable unless outer conditions pass.



## 3. Syntax

```python
if outer_condition:
    if inner_condition1:
        # deepest block
        pass
    elif inner_condition2:
        pass
    else:
        pass
elif another_outer:
    pass
else:
    pass
```

- Indentation level increases with each nesting.
- You can nest as deeply as needed (but shouldn't).
- Ternary can also be nested (avoid for readability):
  ```python
  result = "A" if score >= 90 else "B" if score >= 80 else "C"
  ```

**PEP 8**: Keep nesting shallow. Maximum recommended depth is usually 2–3 levels.



## 4. Machine Learning Usage

- **Data preprocessing pipelines**: Outer check for dataset type, inner feature-specific cleaning.
- **Model training loops**: Early stopping with multiple criteria (outer: epoch limit, inner: patience).
- **Inference routing**: `if model_type == "ensemble":` then inner logic per sub-model.
- **Pandas**: Nested `np.where` or `apply` with conditionals.
- **PyTorch**: Dynamic model behavior (e.g., different forward passes based on training mode + input shape).
- **LLM agents**: Outer intent detection, then nested tool selection logic.

Example in Scikit-learn preprocessing:
```python

    if df['column'].nunique() < 10:
        # one-hot
    else:
        # label or target encode
```

## 5. Interview Questions

## Basic Level

### 1. What is nested `if` and when is it useful?

A nested `if` is simply an `if` statement placed inside another `if` block. It is useful when a second condition only matters **after** a primary condition has already evaluated to true.

* **Example:** Checking if a user is logged in first, and *only then* checking if they have admin privileges.



### 2. Can you access outer variables inside nested `if`?

**Yes.** Variables defined in an outer scope are completely accessible within inner nested blocks.

* **Note:** In Python, variables created inside an `if` block also stay in the outer function scope, but in block-scoped languages like C++ or Java, variables declared inside an inner block are restricted to that block.



### 3. What happens if an inner `if` condition is never reached?

If the outer `if` condition evaluates to `False` (or `false`), the program execution skips the entire inner block. The code inside the inner `if` is never parsed or executed, and any side effects inside it will not happen.

### 5. How does indentation affect nested blocks?

In languages like Python, **indentation defines code blocks**. Missing or inconsistent indentation will cause a `SyntaxError` or alter the program's logic. In curly-brace languages (like JavaScript or Java), indentation doesn't change execution, but proper formatting is critical for readability.



## Intermediate Level

### 1. How would you reduce nesting in a validation function?

You reduce nesting by using **Guard Clauses** (early returns) or combining independent checks. Instead of nesting logic deeper, check for invalid states first, handle them, and return early.

```python
# Deeply nested (Hard to read)
def process_user(user):
    if user:
        if user.is_active:
            if user.has_permission:
                return "Access Granted"

# Guard Clauses (Clean & flat)
def process_user_clean(user):
    if not user:
        return "No user found"
    if not user.is_active:
        return "Account inactive"
    if not user.has_permission:
        return "Permission denied"
        
    return "Access Granted"

```



### 2. Compare nested `if` vs logical `and` chaining.

* **Nested `if`:** Best when you need to perform an operation or `else` branch at an intermediate step before checking the next condition.
* **Logical `and`:** Best when you need **all** conditions to be true before executing a single action. It takes advantage of **short-circuit evaluation** (stops evaluating as soon as one condition fails).



### 3. When should you use nested conditionals vs separate functions?

* **Use nested conditionals:** For small, local logic with 2–3 decisions closely tied together.
* **Use separate functions:** When nested blocks exceed 3 levels deep ("Arrow Anti-pattern"), when an inner block contains complex domain logic, or when a block can be reused elsewhere.



### 4. Explain impact on code coverage and testing.

Every nested `if` creates a new execution path.

* **Branching complexity:** Nesting increases the total number of test cases required to achieve 100% branch coverage exponentially.
* **Hidden bugs:** Deeply nested code makes edge cases harder to visualize, making it easy to miss untested paths in unit tests.



### 5. Refactor this nested code.

def get_discount(user, cart_total):

## Advanced Level

### 1. Design a decision tree with nested conditionals vs Strategy pattern.

* **Nested Conditionals:** Hardcodes rules directly into control flow. Hard to extend—adding a new rule requires modifying existing code, violating the **Open/Closed Principle**.
* **Strategy Pattern:** Encapsulates each rule into its own class or function object. The main runner iterates through a collection of strategies or uses a lookup table (dictionary/map), allowing you to add new behavior without touching core logic.



### 2. Discuss cyclomatic complexity and how nesting affects it.

**Cyclomatic complexity** measures the number of linearly independent paths through a program's source code.

* Every `if`, `elif`, `for`, or `while` loop increases the score by 1.
* Nesting compounds cognitive load because developers must mentally keep track of multiple active context states simultaneously. High cyclomatic complexity (>10) strongly correlates with higher bug rates and lower maintainability.



### 3. How do you handle deeply nested business rules in large codebases?

Engineers use several structural techniques to eliminate deep nesting in enterprise apps:

1. **Rule Engines:** Move decision trees to declarative configuration or specialized rule tools.
2. **Lookup Tables / Dictionaries:** Replace nested branches with key-value mappings to handler functions.
3. **Chain of Responsibility Pattern:** Pass request objects through a chain of handler objects until one processes it.
4. **Pattern Matching (`match/case` in Python 3.10+):** Express complex nested structural checks cleanly.



### 4. Compare Python's approach to nested logic with pattern matching in other languages.

Historically, Python relied heavily on `if/elif/else` or dictionary lookups. Modern Python (3.10+) introduced structural pattern matching (`match/case`), which allows destructuring objects, checking types, and evaluating guards in a single, flat construct—similar to functional languages like Rust, Scala, or Elixir.

```python
# Structural Pattern Matching (Flat & clear)
match command.split():
    case ["quit"]:
        exit()
    case ["load", filename]:
        load_file(filename)
    case ["save", filename] if is_admin:  # Guard condition
        save_file(filename)
    case _:
        print("Unknown command")

```



### 5. Implement a simple parser or router using controlled nesting.

Below is a simple URL route matcher that safely parses top-level sections and handles sub-routes with single-level controlled nesting:

```python
def route_request(path):
    parts = [p for p in path.strip("/").split("/") if p]
    
    if not parts:
        return "Home Page"
        
    endpoint = parts[0]
    
    # First level: Top-level domain routing
    if endpoint == "api":
        if len(parts) > 1 and parts[1] == "v1":
            return "API Version 1 Endpoint"
        return "API Base Endpoint"
        
    if endpoint == "users":
        if len(parts) > 1:
            return f"User Profile for ID: {parts[1]}"
        return "User List"
        
    return "404 Not Found"

# Examples
print(route_request("/api/v1"))   # API Version 1 Endpoint
print(route_request("/users/42"))  # User Profile for ID: 42

```
