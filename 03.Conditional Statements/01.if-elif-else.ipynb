# Conditional Statements (if, elif, else)

## Intuition

Think of conditional statements like a **security guard at an office building**:

- `if` → "If you have an employee badge, let them in."
- `elif` → "Else if you have a visitor pass, escort them to reception."
- `else` → "Otherwise, deny entry."

The guard evaluates conditions **in order** and takes the first matching action, ignoring the rest. This sequential evaluation is key to understanding control flow.


## Syntax

```python
# Basic if
if condition:
    # code block (must be indented)

# if-elif-else chain
if condition1:
    # block 1
elif condition2:
    # block 2
elif condition3:
    # block 3
else:
    # fallback block

# One-line (ternary) - use sparingly
value = true_value if condition else false_value
```

**Important rules**:
- Condition must evaluate to `bool` (truthy/falsy values are allowed but be explicit in production).
- Colon `:` is mandatory.
- Consistent indentation (4 spaces per PEP 8).
- You can nest conditionals, but deep nesting (>3 levels) usually signals a need for refactoring.


## Internal Working

Python evaluates conditions **lazily** from top to bottom. As soon as one condition is `True`, it executes that block and **skips all remaining `elif`/`else`**.

- Each condition is evaluated in its own scope for short-circuiting (e.g., `and`/`or`).
- The code blocks are separate execution frames.
- Boolean context: `0`, `None`, `""`, `[]`, `{}` etc. are falsy. Everything else is truthy.

In CPython, this is implemented via bytecode instructions like `POP_JUMP_IF_FALSE`.


## Built-in Methods / Operations

Conditionals themselves don't have methods, but they work with:

- **Comparison operators**: `==`, `!=`, `>`, `<`, `>=`, `<=`, `is`, `is not`, `in`, `not in`
- **Logical operators**: `and`, `or`, `not` (with short-circuiting)
- `bool()` conversion



````python
# example

age = 25
has_ticket = True

if age >= 18 and has_ticket:
    print("Entry Granted")
elif age < 18:
    print("Minor - parent supervision required")
else:
    print("No Entry")
````

## Common Mistakes

**Mistake 1**: Using assignment `=` instead of comparison `==`


````python
x = 5 # right use

if x = 5: # use '==' insted of '=' : SyntaxError
    print("yes")
````

**Mistake 2**: Deep nesting instead of flat structure


````python
if user:
    if user.is_active:
        if user.has_permission:
            # too deep!
````


**Mistake 3**: Checking `is True` / `== True` instead of truthy

````python
if status == True:          # Fragile
if status is True:                  # Better
````

- Keep conditions **simple and readable** — extract complex logic into well-named functions.
- Use `elif` for mutually exclusive conditions.
- Prefer **guard clauses** (early returns) over deep nesting.
- Avoid magic numbers/strings — use constants or Enums.
- Be explicit: `if len(items) > 0` → `if items:` is acceptable but document intent.
- Use `match` statement (Python 3.10+) for complex structural pattern matching when appropriate.

### Basic

**1. What is the difference between `if` and `elif`?**

The `if` statement is the entry point for any conditional block. It always gets evaluated. `elif` (else if) is only checked if the previous `if` or `elif` condition was False. 

So `if` can stand alone, but `elif` always needs a preceding `if`. It helps avoid checking unnecessary conditions and keeps the logic clean.

**2. What values are considered falsy in Python?**

Python considers these as falsy:
- `False`
- `0`, `0.0`, `0j` (any numeric zero)
- `None`
- Empty sequences: `""`, `[]`, `()`
- Empty mappings: `{}`
- Objects where `__bool__()` returns False or `__len__()` returns 0

Everything else is truthy. I usually use `if my_list:` instead of `if len(my_list) > 0:` because it's more Pythonic.

**3. Write a simple program to check if a number is positive, negative, or zero.**

```python
def check_number(num):
    if num > 0:
        return "Positive"
    elif num < 0:
        return "Negative"
    else:
        return "Zero"

# Example
print(check_number(5))    # Positive
print(check_number(-3))   # Negative
print(check_number(0))    # Zero
```

Simple and straightforward.

**4. Explain short-circuit evaluation with `and`/`or`.**

Short-circuiting means Python stops evaluating as soon as it knows the final result.

- For `and`: if the first condition is False, it doesn't check the second (because False and anything = False).
- For `or`: if the first condition is True, it skips the rest (True or anything = True).

Example:
```python
if user and user.is_active:  # won't call is_active if user is None
    ...
```

This is very useful to prevent errors like AttributeError or division by zero.

**5. What happens if all conditions in an `if-elif-else` are False?**

The `else` block executes. If there's no `else`, then nothing happens — the code just continues after the conditional block.

**3. Compare `if-elif` chain vs `match` statement.**

`if-elif` is very flexible and works everywhere. `match` (structural pattern matching) is much cleaner when you're matching against specific values, enums, or data structures.

```python
# match example
match status:
    case "pending":
        ...
    case "approved":
        ...
    case _:
        ...
```

`match` is more readable for many discrete cases, but `if-elif` is better when you have complex boolean conditions or range checks.

**4. How do conditionals affect code coverage in testing?**

More branches = more test cases needed to achieve high coverage. Each `if`/`elif` path is a separate branch. I make sure to write tests for:
- Every individual condition being True/False
- Edge cases
- Combinations when conditions overlap

Tools like coverage.py help identify missed branches.

### Advanced

**1. Design a state machine using conditionals (or discuss when to use classes instead).**

For simple cases, I can use a dictionary of states and transition functions. But once there are many states or complex behavior, I prefer a proper state pattern with classes.

Conditionals work fine for 4-5 states, but beyond that, the class-based approach is more maintainable.

**2. How do you handle complex business rules without massive `if` chains?**

I avoid big `if` chains by using:
- Rules engine / decision table
- Strategy pattern
- Policy objects
- Sometimes a DSL (Domain Specific Language) if rules change frequently
- Feature flags for toggling rules

Example: I might have a list of Rule objects, each with a condition function and an action.

**3. Discuss performance implications of condition ordering.**

I order conditions from most likely to least likely (or cheapest to most expensive to evaluate). 

For example, check `None` or empty first before doing expensive operations like database calls or regex.

This can give meaningful performance gains in hot paths.

**4. Implement a simple rules engine.**

```python
class Rule:
    def __init__(self, condition, action, priority=0):
        self.condition = condition
        self.action = action
        self.priority = priority

class RulesEngine:
    def __init__(self):
        self.rules = []
    
    def add_rule(self, rule):
        self.rules.append(rule)
        self.rules.sort(key=lambda r: r.priority, reverse=True)
    
    def evaluate(self, data):
        for rule in self.rules:
            if rule.condition(data):
                return rule.action(data)
        return None
```

Then I can define business rules separately and add them easily.

**5. How would you unit test a function with many conditional branches?**

I use pytest parametrize to test each path individually. I also test combinations when conditions interact.

I mock dependencies and focus on:
- Input → Expected output for each branch
- That certain side effects happened (or didn't)
- Error cases

I aim for 100% branch coverage on critical functions.
